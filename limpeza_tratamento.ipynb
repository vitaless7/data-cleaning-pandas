### *Outliers*

```python
df_sem_nulos.describe()
```

```python
import seaborn as sns
```

```python
sns.boxplot(x=df_sem_nulos['cliente.tempo_servico'])
```

```python
Q1 = df_sem_nulos['cliente.tempo_servico'].quantile(.25)
Q3 = df_sem_nulos['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5*IQR
limite_superior = Q3 + 1.5*IQR
```

```python
outliers = (df_sem_nulos['cliente.tempo_servico'] < limite_inferior) | (df_sem_nulos['cliente.tempo_servico'] > limite_superior)
```

```python
df_sem_nulos[outliers]['cliente.tempo_servico'] 
```

### *Substituindo Outliers*

```python
df_sem_out = df_sem_nulos.copy()
```

```python
df_sem_out.describe()
```

```python
df_sem_out.loc[outliers, 'cliente.tempo_servico'] = np.ceil(
    df_sem_out.loc[outliers, 'conta.cobranca.Total']/
    df_sem_out.loc[outliers, 'conta.cobranca.mensal']
)
```

```python
sns.boxplot(x= df_sem_out['cliente.tempo_servico'])
#o número de outliers diminuiu, mas ainda não podemos considerar viável para
#um modelo pre-machine learning
```

### *Removendo Outliers*

```python
Q1 = df_sem_out['cliente.tempo_servico'].quantile(.25)
Q3 = df_sem_out['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5*IQR
limite_superior = Q3 + 1.5*IQR
outliers = (df_sem_out['cliente.tempo_servico'] < limite_inferior) | (df_sem_out['cliente.tempo_servico'] > limite_superior)
```

```python
outliers
```

```python
df_sem_out[outliers]
```

```python
df_sem_out= df_sem_out[~outliers]
df_sem_out
```

```python
sns.boxplot(x=df_sem_out['cliente.tempo_servico'])
```

```python
df_sem_out.reset_index(drop=True,inplace=True)
```

```python
df_sem_out
```
